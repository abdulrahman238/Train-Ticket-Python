**Loading Data**

```python
import pandas as pd 
df=  pd.read_csv(r"C:\Users\User\Desktop\Data Analyst projects Projects others\UK+Train+Rides\railway.csv")
df.head(10)
```

```python
#df.drop_duplicates()
df
```

```python
df["Railcard"] = df["Railcard"].fillna("Unknown")
df["Reason for Delay"]=df["Reason for Delay"].fillna("Unknown")
df
```

```python
df["Date of Journey"] = pd.to_datetime(df["Date of Journey"], errors="coerce")
df["Purchased_Mouth"] = df["Date of Journey"].dt.month_name()
df["Purchased_day"] = df["Date of Journey"].dt.day_name()
df["Purchased_Year"] = df["Date of Journey"].dt.year

# renaming the columns 

df = df.rename(columns={"Purchased_Mouth": "Month_Journey"})
df.rename(columns={"Purchased_day": "Day_Journey"})
df.head()
```

```python
df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"], errors="coerce")
df
```

```python
# converting the date time

df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"], errors="coerce")
df["Month_Purchased"] = df["Date of Purchase"].dt.month_name()
df["Day_Purchased"] = df["Date of Purchase"].dt.day_name()
df["Year_Purchased"] = df["Date of Purchase"].dt.year
df
```

**Show KPIs**

```python
Total_Revenue = df["Price"].sum()
print(Total_Revenue)
Total_Journey = len(df)
print(Total_Journey)
avg_price = int(round(df["Price"].mean(),2))
print(avg_price)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))  

# Properly spaced y-coordinates
plt.text(0.05, 0.7, f"Total Revenue\n${Total_Revenue:,}", fontsize=16, weight='bold', color='green', ha='left')
plt.text(0.05, 0.45, f"Avg Price\n${avg_price:,}", fontsize=16, weight='bold', color='orange', ha='left')
plt.text(0.05, 0.2, f"Total Journey\n{Total_Journey:,}", fontsize=16, weight='bold', color='blue', ha='left')
plt.axis("off")
plt.show()
```

```python
# plotttng pie chart 
Journey_counts = df["Journey Status"].value_counts()
plt.figure(figsize = (6,8))
plt.pie(Journey_counts,
       autopct='%1.1f%%',
          pctdistance=0.8,
          startangle=90,
        textprops={"fontsize": 14, "weight": "bold"},
    
)
plt.title("Journey Status Distribution", fontsize = 16,weight="bold")
plt.show()
```

```python
# Count journeys per month
Month_Journey = df["Month_Journey"].value_counts()  # <-- correct

# Plot bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.bar(Month_Journey.index, Month_Journey.values, color='blue')

plt.title("Number of Journeys per Month", fontsize=16)
plt.xlabel("Month", fontsize=14)
plt.ylabel("Number of Journeys", fontsize=14)
plt.show()
```
